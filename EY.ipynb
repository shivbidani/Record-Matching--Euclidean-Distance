```python
import pandas as pd
import sklearn
import numpy as np
#read file
df1=pd.read_csv("EY1.csv")
df2=pd.read_csv("EY2.csv")
```

```python
def convert_to_days(date):
    su = date.split('/')
    return int(su[0]) + (int(su[1])*30) + (int(su[2])*365)
    #return int(st[0]) + (int(st[1])*30)

#convert date to numerical value for distance calculation

for i in range(0,len(df1)):
    df1['Trade_date'][i] = convert_to_days(df1['Trade_date'][i])
    df1['Settle_date'][i] = convert_to_days(df1['Settle_date'][i])

#convert date to numerical value for distance calculation
for i in range(0,len(df2)):
    df2['Trade_date'][i] = convert_to_days(df2['Trade_date'][i])
    df2['Settle_date'][i] = convert_to_days(df2['Settle_date'][i])

```

```python
dataset_array1 = df1.values

dataset_array2 = df2.values

sm1= np.delete(dataset_array1, [0,1,4,5,6,7,8,10], axis=1)
print(sm1.dtype)
print(sm1)
sm2= np.delete(dataset_array2, [0,1,4,5,6,7,8,10], axis=1)
print(sm2.dtype)
print(sm2)

a = np.array(sm1)
b = np.array(sm2)
dist = np.linalg.norm(a-b)
print(np.array(dist))

a = np.array(sm1)
b = np.array(sm2)
dist = np.linalg.norm(a-b)


#Euclidean distance function
def cal_distance(x,y):
    return np.sqrt(np.dot(x, x) - 2 * np.dot(x, y) + np.dot(y, y))
    

dist = []
for i in a:
    each_dist = []
    for j in b:
        each_dist.append(cal_distance(i, j))
    dist.append(each_dist)

```
